```python
# https://www.acmicpc.net/problem/1991

class Node:
    def __init__(self, value):
        self.value=value
        self.left = None
        self.right = None
```

```python
class graph():
    def __init__(self):
        self.root = None
    
    def find(self, value):
        now = self.root
        will_visit = [now]
        visited = []
        while len(will_visit)>0:
            if now.value==value:
                return now
            else:
                if now.left!=None:
                    will_visit.append(now.left)
                if now.right!=None:
                    will_visit.append(now.right)
                now = will_visit.pop(1)
        return None
    
    def add(self, value, left, right):
        if self.root==None:
            temp = Node(value)
            if left!='.':
                temp.left = Node(left)
            if right!='.':
                temp.right = Node(right)
            self.root = temp
        else:
            now = self.find(value)
            if now!=None:
                if left!='.':
                    now.left= Node(left)
                if right != '.':
                    now.right = Node(right)       
                    
    def search1(self):
        will_visit=[self.root]
        visited = []
        while len(will_visit) > 0:
            now = will_visit.pop()
            if now.right!=None:
                will_visit.append(now.right)
            if now.left!=None:
                will_visit.append(now.left)
            visited.append(now)
        for node in visited:
            print(node.value,end='')
        
    def search2(self):
        checked = []
        will_visit = [self.root]
        visited = []
        while len(will_visit) >0:
            now = will_visit.pop()
            # 현재 노드가 leaf 노드이면 바로 visit
            if now.left==None and now.right==None:
                visited.append(now)
            # 아닐때
            else:
                # 이미 check 했었나 확인
                flag = True
                for item in checked:
                    if item==now:
                        flag=False
                        break
                # 아직 check 되지 않았을때
                if flag:
                    # 오른쪽 넣고
                    if now.right!=None:
                        will_visit.append(now.right)
                    # 자기 넣고
                    will_visit.append(now)
                    # 왼쪽 넣고
                    if now.left!=None:
                        will_visit.append(now.left)
                    # check 했다고 넣고
                    checked.append(now)
                # 이미 check 한 노드일때
                else:
                    # 바로 방문
                    visited.append(now)
        for node in visited:
            print(node.value, end='')
                    
    def search3(self):
        checked = []
        will_visit = [self.root]
        visited = []
        while len(will_visit) >0:
            now = will_visit.pop()
            # 현재 노드가 leaf 노드이면 바로 visit
            if now.left==None and now.right==None:
                visited.append(now)
            # 아닐때
            else:
                # 이미 check 했었나 확인
                flag = True
                for item in checked:
                    if item==now:
                        flag=False
                        break
                # 아직 check 되지 않았을때
                if flag:
                    # 자기 넣고
                    will_visit.append(now)
                    # 오른쪽 넣고
                    if now.right!=None:
                        will_visit.append(now.right)
                    # 왼쪽 넣고
                    if now.left!=None:
                        will_visit.append(now.left)
                    # check 했다고 넣고
                    checked.append(now)
                # 이미 check 한 노드일때
                else:
                    # 바로 방문
                    visited.append(now)
        for node in visited:
            print(node.value, end='')                
```

```python
n = int(input())
test_graph = graph()
for _ in range(n):
    value, left, right = map(str, input().split(' '))
    test_graph.add(value, left, right)
test_graph.search1()
print()
test_graph.search2()
print()
test_graph.search3()

# 풀긴 풀었는데 이거 뭔가 문제 있다
```

```python
# 선생님
# 이거 굉장히 많이 나오는 기본적인 유형이니까 꼭 기억하자.
# 전위순회, 중위순회, 후위순회는 알아두자.

# 트리 순회 재귀적으로도 구현 많이함. 오..
# 이거 꿀팁이네
# 재귀함수로 구현하면 구현 개꿀이네 물론 큰 트리 같은 경우 힘들겠지만

# 중위 순회는 또 재밌는 면을 가지고 있는데
# 중위 순회의 결과는 트리를 아래로 꾹 눌렀을때 나오는 순서랑 같음
# 아 다음 문제 이 중위순회 속성이랑 BFS랑 섞어서 쓰면 쉬웠겠ㄴ네

# 코딩테스트에선 클래스가 크면 별로임

class Node:
    def __init__(self, data, left_node, right_node):
        self.data = data
        self.left_node = left_node
        self.right_node = right_node
        
def pre_order(node):
    print(node.data, end='')
    if node.left_node != '.':
        pre_order(tree[node.left_node])
    if node.right_node != '.':
        pre_order(tree[node.right_node])

def in_order(node):
    if node.left_node != '.':
        in_order(tree[node.left_node])
    print(node.data, end='')
    if node.right_node != '.':
        in_order(tree[node.right_node])

def post_order(node):
    if node.left_node != '.':
        post_order(tree[node.left_node])
    if node.right_node != '.':
        post_order(tree[node.right_node])
    print(node.data, end='')

n = int(input())
tree = {}

for i in range(n):
    data, left_node, right_node = input().split()
    tree[data] = Node(data, left_node, right_node)

pre_order(tree['A'])
print()
in_order(tree['A'])
print()
post_order(tree['A'])
```

```python
# https://www.acmicpc.net/problem/2250

# 두 노드 사이 거리 구하는 법 -> 중위순회로 눌러놓고 구하면 됨

# 각 레벨별 끝 노드 구하는 법
# 레별별로 딕셔너리 만듬
# BFS 하는데 BFS넣는 곳을 레벨별로 배열에 넣음.
# 딕셔너리에서 배열 하나씩 꺼내서 배열의 0번째 인덱스가 가장 왼쪽놈 -1 인덱스가 가장 오른쪽놈

graph = {}
n = int(input())
for _ in range(n):
    value, left, right = map(int, input().split(' '))
    graph[value] = [left, right]
    
inorder_arr = []

def inorder(start):
    if graph[start][0] != -1:
        inorder(graph[start][0])
    inorder_arr.append(start)
    if graph[start][1] != -1:
        inorder(graph[start][1])
        
inorder(1)

bfs_level = {}

def bfs_per_level(start, level):
    if bfs_level.get(level)!=None:
        bfs_level[level].append(start)
    else:
        bfs_level[level] = [start]
    if graph[start][0]!=-1:
        bfs_per_level(graph[start][0], level+1)
    if graph[start][1]!=-1:
        bfs_per_level(graph[start][1], level+1)
        
bfs_per_level(list(graph.keys())[0],1)

max_len = 0
max_level = 0
for arr in bfs_level:
    start = bfs_level[arr][0]
    end = bfs_level[arr][-1]
    length_now = inorder_arr.index(end) - inorder_arr.index(start)+1
    if length_now > max_len:
        max_level = arr
        max_len = length_now
        
print(max_level, max_len)

# 논리적으론 맞는거 같은데 런터임 에러 난다.
```

```python
# 선생님

# 상당히 난이도가 있는문제. 중으로 써있는데 상 정도임
# 구현하는데 실수 할 만한 구석이 많음
# 트리의 특징을 잘 알아야 함.

# 전 문제에서 말한 것 처럼 중위순회를 잘 이용하되 레벨을 잘 처리 해야됨.
# 중위순회 하면서 레벨 값을 기록해뒀다가 너비를 구하면 됨.
# 이 정도면 코딩테스트에서 답이 긴편인거 같음.

# 아 이 문제에선 처음받은 값이 항상 루트는 아님.
# 루트 찾는것부터 해야됨.따라서 노드에 parent도 기록해 두어야 함.

# 오 이건 구현을 좀 특이하게 하셨네
# 트리의 key엔 그냥 index를 넣어놓고 Node에 data, parent, left, right 다 넣음

# 알고리즘의 복잡도 같은 면에선 그렇게 어렵진 않지만
# 구현 부분이 어려운 문제임

class Node:
    def __init__(self, number, left_node, right_node):
        self.parent = -1
        self.number = number
        self.left_node = left_node
        self.right_node = right_node
        
def in_order(node, level):
    global level_depth, x
    level_depth = max(level_depth, level)
    if node.left_node != -1:
        in_order(tree[node.left_node], level + 1)
    level_min[level] = min(level_min[level], x)
    level_max[level] = max(level_max[level], x)
    x += 1
    if node.right_node != -1:
        in_order(tree[node.right_node], level + 1)

n = int(input())
tree = {}
level_min = [n]
level_max = [0]
root = -1
x = 1
level_depth = 1

for i in range(1, n + 1):
    tree[i] = Node(i, -1, -1)
    level_min.append(n)
    level_max.append(0)

for _ in range(n):
    number, left_node, right_node = map(int, input().split())
    tree[number].left_node = left_node
    tree[number].right_node = right_node
    if left_node != -1:
        tree[left_node].parent = number
    if right_node != -1:
        tree[right_node].parent = number

for i in range(1, n + 1):
    if tree[i].parent == -1:
        root = i
        
in_order(tree[root], 1)
result_level = 1
result_width = level_max[1] - level_min[1] + 1

for i in range(2, level_depth + 1):
    width = level_max[i] - level_min[i] + 1
    if result_width < width:
        result_level = i
        result_width = width
    
print(result_level, result_width)
```
