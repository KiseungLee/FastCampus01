```python
# https://www.acmicpc.net/problem/1668

n = int(input())

array = []

for _ in range(n):
    array.append(int(input()))

left=0
right=0

left_max = 0
right_max = 0

for i in range(n):
    if left_max<array[i]:
        left+=1
        left_max = array[i]
        
for i in range(n-1, -1, -1):
    if right_max < array[i]:
        right+=1
        right_max = array[i]
        
print(left)
print(right)
```

```python
# 선생님

# 역시 갯수가 50까지 밖에 없으니 아무렇게나 풀어도 됨 ㅋㅋ
# 여기선 한쪽에서 봤을때 보이는 거 갯수 찾는 함수를 하나 만들고 
# 입력받은 어레이에서 한번 돌리고, 뒤집어서 한번 돌리고

# 뒤집을땐 array.reverse() 를 알아두자.

def ascending(array):
    now = array[0]
    result = 1
    for i in range(1, len(array)):
        if now < array[i]:
            result += 1
            now = array[i]
    return result


n = int(input())
array = []

for _ in range(n):
    array.append(int(input()))
    
print(ascending(array))
array.reverse()
print(ascending(array))
```

```python
# https://www.acmicpc.net/problem/1236

n, m = map(int, input().split(' '))


array_of_array = []
for i in range(n):
    array_of_array.append(list(map(str, input().split(' '))))

    
rows_num = 0
for i in range(n):
    if 'X' not in array_of_array[i]:
        rows_num+=1

        
cols_num = m
for i in range(m):
    for j in range(n):
        if array_of_array[j][i]=='X':
            cols_num-=1
            break

        
print(max(rows_num, cols_num))

# 아이디어는 X가 없는 rows, X가 없는 cols의 갯수중 더 큰걸 뽑으면 됨.
# 아니 런타임 에러가 어디서 나냐구요
```

```python
# 선생님

# 아이디어는 나랑 같은데 내가 아직 배열을 다루는 솜씨가 부족하구나...

n, m = map(int, input().split())
array = []

for _ in range(n):
    array.append(input())
    
row = [0] * n
column = [0] * m

for i in range(n):
    for j in range(m):
        if array[i][j] == 'X':
            row[i] = 1
            column[j] = 1

row_count = 0
for i in range(n):
    if row[i] == 0:
        row_count += 1

column_count = 0
for j in range(m):
    if column[j] == 0:
        column_count += 1

print(max(row_count, column_count))
```
