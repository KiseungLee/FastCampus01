```python
# https://www.acmicpc.net/problem/1991

n = int(input())
dic = {}

for _ in range(n):
    node, left, right = map(str, input().split())
    dic[node] = (left, right)
    
def preorder(node):
    if node!='.':
        path.append(node)
    if node in dic:
        preorder(dic[node][0])
        preorder(dic[node][1])
        
def inorder(node):
    if node in dic:
        inorder(dic[node][0])
    if node!='.':
        path.append(node)        
    if node in dic:
        inorder(dic[node][1])        
        
def postorder(node):
    if node in dic:
        postorder(dic[node][0])
        postorder(dic[node][1])
    if node!='.':
        path.append(node)       
        
path=[]
preorder('A')
print(''.join(path))

path=[]
inorder('A')
print(''.join(path))

path=[]
postorder('A')
print(''.join(path))
```

```python
# 선생님
# 이거 굉장히 많이 나오는 기본적인 유형이니까 꼭 기억하자.
# 전위순회, 중위순회, 후위순회는 알아두자.

# 트리 순회 재귀적으로도 구현 많이함. 오..
# 이거 꿀팁이네
# 재귀함수로 구현하면 구현 개꿀이네 물론 큰 트리 같은 경우 힘들겠지만

# 중위 순회는 또 재밌는 면을 가지고 있는데
# 중위 순회의 결과는 트리를 아래로 꾹 눌렀을때 나오는 순서랑 같음
# 아 다음 문제 이 중위순회 속성이랑 BFS랑 섞어서 쓰면 쉬웠겠ㄴ네

# 코딩테스트에선 클래스가 크면 별로임

class Node:
    def __init__(self, data, left_node, right_node):
        self.data = data
        self.left_node = left_node
        self.right_node = right_node
        
def pre_order(node):
    #################################################
    print(node.data, end='')  # 바로 print 찍어버림
    #################################################
    if node.left_node != '.':
        pre_order(tree[node.left_node])
    if node.right_node != '.':
        pre_order(tree[node.right_node])

def in_order(node):
    if node.left_node != '.':
        in_order(tree[node.left_node])
    print(node.data, end='')
    if node.right_node != '.':
        in_order(tree[node.right_node])

def post_order(node):
    if node.left_node != '.':
        post_order(tree[node.left_node])
    if node.right_node != '.':
        post_order(tree[node.right_node])
    print(node.data, end='')

n = int(input())
tree = {}

for i in range(n):
    data, left_node, right_node = input().split()
    tree[data] = Node(data, left_node, right_node)

pre_order(tree['A'])
print()
in_order(tree['A'])
print()
post_order(tree['A'])
```
