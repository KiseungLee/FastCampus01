```python
# https://www.acmicpc.net/problem/1874

# 문제 설명이 좀 그렇다.

n = input()
numbers = list([input() for i in range(0, int(n))])
numbers2 = numbers.copy()
# 일단 첫번째 수까지는 계속 push
# 첫번째 수를 pop
# 두번째 수가 첫번째 수보다 크다면 거기까지 push
# 그리고 바로 pop
stack = list()
pop_list = list()
return_list = list()

# 잎으로 딱 16턴만 돌거임
num = 0
push_count = 0
pop_count = 0
numbers_index = 0
while push_count<int(n) or pop_count<int(n):
    # 턴의 시작은 push_count가 n인지 확인후 
    # 8이면 numbers 있는걸 pop함
    if push_count==n:
        pop_list.append(stack[-1])
        del stack[-1]
        pop_count+=1
        return_list.append('-')
    # 아니라면 numbers의 가장 앞이 num보다 큰지 확인
    else:
        # 같거나 작으면 pop
        if int(numbers[0]) == num or int(numbers[0]) < num :
            pop_list.append(stack[-1])
            del stack[-1]
            pop_count+=1
            del numbers[0]
            return_list.append('-')
        # 크다면 push
        elif int(numbers[0]) >= num:
            num+=1
            stack.append(num)
            push_count+=1
            return_list.append('+')
            
# 다 돌았으면 popStack이랑 numbers랑 같은지 확인
ok = True
for i in range(len(pop_list)):
    if pop_list[i] != int(numbers2[i]):
        ok = False

if ok:
    for i in return_list:
        print(i)
else:
    print("NO")
    
# 풀긴 했는데 너무 지저분쓰...

# 정리 가능할 거 같음.
# 수열 만들 수 없으면 중간에 깨치는게 좋을거 같음.
```

```python
# 선생님 풀이
# 깰끔쓰.. 일단 입력 받는거 자체를 나눠서 받아도 됨.

n = int(input())
count = 1
stack = []
result = []

for i in range(1, n + 1): # 데이터 개수만큼 반복
    data = int(input())
    while count <= data: # 입력 받은 데이터에 도달할 때까지 삽입
        stack.append(count)
        count += 1
        result.append('+')
    if stack[-1] == data: # 스택의 최상위 원소가 데이터와 같을 때 출력
        stack.pop()
        result.append('-')
    else: # 불가능한 경우
        print('NO')
        exit(0)
        
print('\n'.join(result)) # 가능한 경우 # 출력도 이거 팁
```
