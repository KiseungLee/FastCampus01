## 이번엔 dp 구성까진 직접 짰는데

## 구성한 dp를 채우는 법을 좀 힌트 받음

```python
def solution(k, travel):
    dp = [[0]*(len(travel)+1) for i in range(k+1)] # dp[i][j] = i시간에 j도시에 있을 때 최대 모금액
    
    # dp 채우는건 한 도시당 시간을 쭉 돌거임.
    # 시간 돌때 자전거로 한번, 도보로 한번 돌거임
    for j in range(1, len(travel)+1):
        for i in range(k+1):
            # 자전거 - j번째 마을 자전거로 가는 시간이 i보다 크면 그 마을은 애초에 못가는 마을
            if travel[j-1][2]>i:
                continue
            else:
                time = travel[j-1][2]   # 자전거로 j번째 마을에 도착하는 시간
                if not (j!=1 and dp[i-time][j-1]==0):
                    dp[i][j] = dp[i-time][j-1] + travel[j-1][3]  # j-1 번째 마을에 i-time 시에 도착했을 때 모금액에 j번째 마을 자전거 모금액 더함 
                
            # 도보
            if travel[j-1][0]>i:
                continue
            else:         
                time = travel[j-1][0]
                if not (j!=1 and dp[i-time][j-1]==0):
                    dp[i][j] = max(dp[i][j], dp[i-time][j-1] + travel[j-1][1] ) # 자전거로 갔을때랑, 도보로 갔을때 중 큰 모금액 선택
        
    return dp[-1][-1]
```

```python
k_ = 1650
travel_ = [[500, 200, 200, 100], [800, 370, 300, 120], [700, 250, 300, 90]]
solution(k_, travel_)
```
