```python
name = 'JAN'
```

```python
A_to_name = [ ord(c) - ord('A') if c <= 'N' else 26 + ord('A') - ord(c) for c in name]
done = [False]*len(name)
done_count = 0
```

```python
A_to_name
```

```python
done
```

```python
# 일단 맨 처음거 바꾸고 시작
idx_now = 0
count = A_to_name[idx_now]
done[idx_now] = True
for idx, n in enumerate(A_to_name):
    if n==0:
        done[idx] = True
done_count += 1

while(done_count < len(name)):
    # 왼쪽으로 갈 때 필요한 조작수
    left = 1
    left_idx = idx_now - 1 if idx_now!=0 else len(name)-1
    # 아직 안된거 만날때 까지 왼쪽으로 감
    while(done[left_idx]):
        print('left : ', done, left_idx, idx_now)
        left_idx = idx_now - 1 if left_idx!=0 else len(name)-1
    left += A_to_name[left_idx]
    
    # 오른쪽으로 갈 때 필요한 조작수
    right = 1
    right_idx = idx_now + 1 if idx_now!=len(name)-1 else 0
    while(done[right_idx]):
        print('right : ', done, right_idx, idx_now)
        right_idx = idx_now + 1 if right_idx!=len(name)-1 else 0
    right += A_to_name[right_idx]
    
    # 짧은 쪽으로 이동
    if left<right:
        print('go left')
        count += left
        done[left_idx] = True
        idx_now = left_idx
    else:
        print('go right')
        count += right
        done[right_idx] = True
        idx_now = right_idx
        
    done_count += 1
    
print(count)
```
