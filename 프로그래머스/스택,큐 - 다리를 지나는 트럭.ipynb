```python
# https://programmers.co.kr/learn/courses/30/lessons/42583?language=python3

bridge_length = 100
weight = 100
truck_weights = [10, 10, 10, 10, 10, 10, 10, 10, 10, 10]
```

```python
from collections import deque

arr = deque([0]*bridge_length) # 다리
```

```python
time = 0 # 지나간 시간
weight_now = 0 # 현재 다리에 올라간 무게
```

```python
while truck_weights or weight_now!=0:
    print(truck_weights, arr,weight_now )
    time += 1
    # 차 나감
    out = arr.popleft()
    if out>0:
        weight_now -= out
    # 차 들어옴
    if truck_weights and weight_now+truck_weights[0] <= weight:
        put = truck_weights.pop(0)
        arr.append(put)
        weight_now += put
    else:
        arr.append(0) # 무게 초과거나 남은차가 없으면 안들어 올수도
```
