```python
# 선생님
# 일단 다익스트라 알고리즘 문제임
# 다익스트라 알고리즘은 다이나믹 프로그래밍으로 분류되기도 하고 그리디 알고리즘으로 분류 되기도 함.
# https://www.youtube.com/watch?v=611B-9zk2o4 선생님 유튜브인데 이게 설명이 좋네

import heapq
import sys
# input = sys.stdin.readline

def dijkstra(start):
    heap_data = []                           # 힘 배열
    heapq.heappush(heap_data, (0, start))    # (시작 초=0, 시작 노드=start)과 넣어줌
    distance[start] = 0                      # 감염시간 담는 배열에 시작지점 감염시간을 0으로
    while heap_data:
        dist, now = heapq.heappop(heap_data) # 배열에서 하나 꺼냄 (도달한 초, 노드)
        if distance[now] < dist:             # 현재 힙에서 꺼내 잡고 있는 노드에 바이러스 도착한 시간이 감염시간 담는 배열에 저장된 감염시간 보다 크면
            continue                        # 패스
        for i in adj[now]:                   # 현재 잡고 있는 노드에 연결된 노드를 순회
            cost = dist+i[1]                 # 현재 잡고 있는 노드에 바이러스 도착시간 + 다음 노드까지 가는데 걸리는 시간 셋팅
            if distance[i[0]] > cost:        # 그 셋팅한게 현재 저장되어 있는 감염시간 배열보다 작으면
                distance[i[0]] = cost        # 감염시간 배열에 더 작은 값으로 바꿔주고
                heapq.heappush(heap_data, (cost, i[0])) # 힙배열에 넣어줌.
                
for _ in range(int(input())):                  # 테스트 케이스
    n, m, start = map(int, input().split())    # 컴퓨터 n개, 의존 m개, 시작 start
    adj = [[] for i in range(n+1)]            # 입접 배열
    distance = [1e9]*(n+1)                     # 감염 시간(최대한 작은게 들어감)
    for _ in range(m):                         # 의존 받기
        x, y, cost = map(int, input().split()) # x <- y, cost 초 걸림
        adj[y].append((x,cost))                # 입력
    dijkstra(start)                            # 다익스트라 돌림.
    count = 0                                  # 감염된 컴퓨터수 세줄 변수
    max_distance = 0                           # 마지막으로 감염된 컴퓨터
    for i in distance:                         # 감염 시간 배열을 돌음
        if i!= 1e9:                            # 초기 설정한 1e9면 감염 안됨.
            count += 1                         # 감염 됐으면 count += 1
            if i>max_distacne:                 # 지금 보고 있는 컴퓨터의 감염시간이 최대 감염시간보다 크면
                max_distance = i               # 그걸 최대 감염시간으로.
    print(count, max_distance)
```
