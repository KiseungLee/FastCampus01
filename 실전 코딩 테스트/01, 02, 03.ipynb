## 03

ㅋㅋ '맞왜틀'이라는 말이 있다고함. '나는 맞았는데 왜 틀렸나요?' 

근데 맞왜틀의 대부분은 니들이 틀린거야~

연습때 맞왜틀 오면 그냥 다시 해보면서 왜 틀렸는지 등등 생각해보면 되는데

실전에서 맞왜틀 오면 멘탈 흔들릴수 있으니 조심할 것.

맞왜틀의 유형 
1. 컴파일 에러 -> 문법오류
2. 시간 초과 -> 최적화 필요
3. 메모리 초과 ->최적화 필요
4. 런타임 에러 -> 과정 오류 (0으로 나누는 경우, index 범위, 무한루프 등등)
5. 틀렸습니다. -> 그냥 틀렸어~

런터임 에러는 정확히 어떤건지 알기 어렵기 때문에 런타임 에러 뜨면 어디가 에러인지를 잘 잡아야 함.

틀렸습니다 뜨면 
1. 조건(제한, 대소관계(이상, 이하, 초과, 미만) 등등) 
2. 예외처리(ex) 답이 없으면 -1을 출력해라) 
3. 입출력 양식(공백, 순서 등..) 
4. 시간제한, 메모리 제한(언어 자체에서 메모리 초과가 아니라 문제에서 아예 걸어둔 거)
5. 알고리즘이 맞는가?
6. 생각한대로 구현이 잘 됐는가
7. 불필요한 반복문이 있는가
8. 중복 처리는 했는가.

위와 같은 것들을 검토해 봐야 함.
