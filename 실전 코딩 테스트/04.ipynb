## 05

구현이란 무엇인가. 이번 강의의 주제임.



파이썬 내장함수 : 외부 모듈 필요없이 사용 가능한 함수.

외부 모듈 : import 쓰는거.



거의 대부분의 코딩테스트 문제는 내장함수만 써도 풀림.

물론 외부 모듈을 쓰면 쉽게 풀리는 경우도 있음.



이 강의에선 시뮬레이션을 하면서 문제를 풀거임. 완전 구체적인 코드까진 안쓴다는 건가.

시뮬레이션 이란
1. 문제 상황과 일대일 매칭
2. 보다 효율적으로 짜기
3. 최적화와 알고리즘 등 복잡한 코드 까지



코딩테스트를 위한 테크닉들도 많이 알아두면 좋음
